```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
data=pd.read_csv("canada_per_capita_income.csv")
data.head()
```

```python
data.info()
data.describe()
data.isnull().sum()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income(US$)")
plt.scatter(data["year"],data["per capita income (US$)"],color="green",marker=".")
plt.show()
```

```python
x=data[["year"]]
y=data["per capita income (US$)"]
```

```python
#Split the dataset
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Train the model
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
print(f"coefficient:{lm.coef_}")
print(f"Intercept:{lm.intercept_}")
```

```python
#Predict the test set 
y_pred=lm.predict(x_test)
for i in y_pred:
    print(i)
```

```python
%matplotlib inline
plt.xlabel("year_test")
plt.ylabel("per capita income prediction(US$)")
plt.scatter(x_test, y_test, color="blue",marker="+")
plt.plot(x_test, lm.predict(x_test), color="pink")
plt.show()
```
