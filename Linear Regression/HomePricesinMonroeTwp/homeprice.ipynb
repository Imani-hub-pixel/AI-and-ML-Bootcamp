```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model 
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price(US$)")
plt.scatter(df["area"], df["price"], color='red', marker='+')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df["price"])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

y_pred=135.78767123*3300+180616.43835616432
print(y_pred)
```

```python
%matplotlib inline
plt.xlabel("area",fontsize=20)
plt.ylabel("price",fontsize=20)
plt.scatter(df["area"],df["price"],color="red",marker="+")
plt.plot(df["area"],reg.predict(df[["area"]]),color="blue")
plt.show()
```

```python
data=pd.read_csv("predhomeprices.csv")
data.head()
```

```python
predict_prices=reg.predict(data)
```

```python
data["price"]=predict_prices
```

```python
data.to_csv("predhomeprices.csv",index=False)
```
