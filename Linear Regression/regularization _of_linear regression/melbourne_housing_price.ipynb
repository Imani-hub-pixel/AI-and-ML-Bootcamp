```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Suppressing warnings for a clean notebook output
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the dataset
df=pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
useful_columns=['Suburb','Rooms','Type','Method','SellerG','Regionname',
                'Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=df[useful_columns]
df.head()
```

```python
df.isna().sum()
```

```python
columns_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']

for col in columns_to_fill_zero:
    df[col].fillna(0,inplace=True)
df.isna().sum()
```

```python
columns_fill_mean=['Landsize','BuildingArea']
for col in columns_fill_mean:
    df[col].fillna(df[col].mean(),inplace=True)
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True).astype(int)
df.head()
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge  
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```
